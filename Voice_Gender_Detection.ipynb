```python
from transformers import pipeline
import torch

# Check if GPU is available
device = 0 if torch.cuda.is_available() else -1
print(f"Using: {'GPU' if device == 0 else 'CPU'}")

# Load the gender classification model
print("Loading model (this may take a minute)...")
gender_classifier = pipeline(
    "audio-classification",
    model="alefiury/wav2vec2-large-xlsr-53-gender-recognition-librispeech",
    device=device
)
print("Model loaded!")
```

```python
import glob
import pandas as pd
from tqdm import tqdm

def classify_gender_batch(audio_folder, group_name):
    """
    Classify gender for all audio files in a folder.

    Parameters:
        audio_folder: Path to folder containing .wav files
        group_name: Label for this group (e.g., 'HC' or 'PD')

    Returns:
        DataFrame with results
    """
    audio_files = glob.glob(f'{audio_folder}/*.wav')
    print(f"Found {len(audio_files)} files in {audio_folder}")

    results = []

    for audio_file in tqdm(audio_files, desc=f"Processing {group_name}"):
        try:
            # Get prediction
            prediction = gender_classifier(audio_file)

            # Extract best prediction
            top_pred = max(prediction, key=lambda x: x['score'])

            results.append({
                'ID': os.path.basename(audio_file),
                'Group': group_name,
                'Gender': top_pred['label'],
                'Confidence': round(top_pred['score'], 4)
            })
        except Exception as e:
            print(f"Error processing {audio_file}: {e}")
            results.append({
                'ID': os.path.basename(audio_file),
                'Gender': 'error',
            })

    return pd.DataFrame(results)
```

```python
# Classify Healthy Control files
hc_results = classify_gender_batch('HC_audio', 'HC')

print("\n=== HC Results ===")
print(hc_results)
print("\n=== HC Gender Distribution ===")
print(hc_results['Gender'].value_counts())
```
